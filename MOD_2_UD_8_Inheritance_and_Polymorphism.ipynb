```python
# Let's see how to define the Food class and how we can use the concept of inheritance on the class itself

class Food:
  """
  An object of class Food is defined with two attributes, type and name
  """
  def __init__(self, type, name):
    # Each example of Food has an associated type and name
    self.type = type
    self.name = name

  # Methods of the Food class
  def tell_me_type(self):
    print("Type:", self.type)

  def tell_me_name(self):
    print("Name:", self.name)


# We create an instance object of the class Food
my_fruit = Food('fruit', 'pear')
my_fruit.tell_me_type()
my_fruit.tell_me_name()

```

```python
## Inheritance
# We define class Fish, daughter of class Food
class Fish(Food):
  pass

# The Fish class inherits the methods of the Food class
my_fish = Fish('fish', 'trout')
my_fish.tell_me_type()
my_fish.tell_me_name()
```

```python
# Polymorphism
class Fish():
  def recommended_intake(self):
    print("It is recommended to eat fish 3 times a week")
class Meat():
  def recommended_intake(self):
    print("Try to limit your meat consumption to less than 3 times a week")

my_fish = Fish()
my_meat = Meat()

my_fish.recommended_intake()
my_meat.recommended_intake()
```
