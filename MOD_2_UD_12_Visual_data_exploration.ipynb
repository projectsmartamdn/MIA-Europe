## Diabetes data set

```python
from sklearn.datasets import load_diabetes
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid") # Show position grid on charts
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None) # This option allows you to display all columns in the dataset
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

"""
In this notebook we will work with the Sci-Kit Learn Diabetes Dataset
https://scikit-learn.org/stable/datasets/toy_dataset.html#diabetes-dataset
And we will use the Seaborn library for data visualization
https://seaborn.pydata.org/
"""

data = load_diabetes()
df = pd.DataFrame(data=data.data, columns=data.feature_names)

print(df.describe())
print("\n", 50 * "*", "\n Dataset Description: \n\n", data.DESCR)
```

```python
# The DESCR method of the dataset allows us to expand the information about the working dataset
# We are going to rename the columns following the nomenclature of the clinical variables
df = df.rename(columns={'s1': 'tc', 's2': 'ldl', 's3': 'hdl', 's4': 'tch', 's5': 'ltg', 's6': 'glu'})
print(df)
# As you will have noticed, this dataset has been previously normalised
# to facilitate the application of certain machine learning models
```

## Scatter plot

```python
# Let's see how LDL levels and Body Mass Index (BMI) are related
fig = plt.figure(1, figsize=(15,5))
fig.add_subplot(121)
sns.scatterplot(data=data, x=df['bmi'], y=df['ldl'], palette='Pastel1', legend='full')
plt.xlabel('BMI')
plt.ylabel('LDL')

# To evaluate the separation between data, we would need a categorical label for
# each variable, let's make a dummy
import numpy as np

class_l = np.random.randint(2, size=len(df['bmi'].values))
df_l = df
df_l['class'] = class_l
fig.add_subplot(122)
sns.scatterplot(data=df_l, x=df_l['bmi'], y=df_l['ldl'], palette='Pastel1', legend='full', hue='class');

# As we can see, having used a random distribution, the data appears mixed
```

## Box plots

```python
# To reduce execution time,
# we reduce the dataset to the following three variables
df_s = df[['age', 'bmi', 'ldl']]
sns.boxplot(data=df_s);
```

## Distribution graphs

```python
from scipy.stats import norm # We use the normal scipy stats function to
# fit the histogram of the distribution

pars = norm.fit(df['bmi'].values)
sns.distplot(df['bmi'], kde=False, fit=norm, fit_kws={'color': 'r', 'linewidth': 2.5})
plt.title("Mean: {:.2f}, Sigma: {:.2f}".format(pars[0], pars[1]))
```

## Pair charts

```python
df_l = df_l[['age', 'bmi', 'ldl', 'class']]
sns.pairplot(df_l, palette='Pastel1', hue='class')
plt.show()
```
