```python
# We can load a module or complete library using 'import'
import skimage
from skimage import data

# The skimage library is composed of different sub-modules, such as data
# The following command allows us to download a histological slice image
# included inside data
ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# To import only the 'data' submodule we can use '.' which in Python indicates <<is part of>>
import skimage
from skimage import data

ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# Another way to import it is as follows
import skimage
from skimage import data

ihc_rgb = data.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# We can even rename the imported module name
from skimage import data as dt

ihc_rgb = dt.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)

# Or in another way
import skimage.data as md

ihc_rgb = md.immunohistochemistry()
print("ihc_rgb shape:", ihc_rgb.shape)
```

```python
# This is the common way to work with the graphics module of the matplotlib library
import matplotlib.pyplot as plt

plt.imshow(ihc_rgb);
```

```python
# The external library pkgutil allows us to list all the modules that make up
# the sci-kit image library
# https://docs.python.org/3/library/pkgutil.html
import pkgutil
import skimage
package=skimage
for importer, modname, ispkg in pkgutil.walk_packages(path=package.__path__,
                                                      prefix=package.__name__+'.',
                                                      onerror=lambda x: None):
    print(modname)
```
